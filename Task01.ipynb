```python
import matplotlib.pyplot as plt

# Generate arrays
exp_data = np.random.exponential(scale=1.0, size=100000)
print(exp_data)

uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

# Plot histograms
plt.hist(exp_data, bins=200, density=True, histtype="step", color="red", label="Exponential")
plt.hist(uni_data, bins=50, density=True, histtype="step", color="blue", label="Uniform")
plt.hist(norm_data, bins=150, density=True, histtype="step", color="green", label="Normal")

plt.xlim(-5, 5)
plt.ylim(0, 1)

plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Z = X² + Y²")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Download the CSV
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset using pandas
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin1')
# Display first 5 rows
df.head()

# Select numeric Pokémon stats for correlation
cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_df = df[cols].dropna()

# Pearson correlation
pearson_corr = stats_df.corr(method="pearson")


# Spearman correlation
spearman_corr = stats_df.corr(method="spearman")

# Pearson heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Spearman heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```
